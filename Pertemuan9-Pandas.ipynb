### 29. Melakukan random sampling pada sebuah data frame

```python
import pandas as pd

print(pd.__version__)
```

```python
d = {'col_1':[1, 2, 3, 4, 5], 
     'col_2':[10, 20, 30, 40, 50]}
df = pd.DataFrame(d)
df
```

```python
df.sample(n=4, replace=False, random_state=0) # Membuat sebuah data frame dengan nilai teracak tanpa ada nilai yang redundant.
```

```python
df.sample(n=4, replace=True, random_state=0) # Membuat sebuah data frame dengan nilai teracak dengan kemungkinan ada nilai yang redundant.
```

## 31. Tipe data ordinal pada pandas.

```python
from pandas.api.types import CategoricalDtype

tingkat_kepuasan = CategoricalDtype(['buruk', 'cukup', 'baik', 'sangat baik'], 
                                    ordered=True)

df['kepuasan'] = df['kepuasan'].astype(tingkat_kepuasan)
df
```

```python
df = df.sort_values('kepuasan', ascending=True)
df
```
