# Pandas 9
## Membagi Data Frame menjadi dua secara acak

```python
import pandas as pd
import numpy as np
# Importing modules numpy and pandas.
```

```python
# Membuat sebuah data frame yang memiliki nilai bilangan bulat random 1 sampai 19 didalamnya.
# Dataframe memiliki ukuran 10 baris dan 5 kolom.
n_rows= 10
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns=cols)
df
```

```python
df.shape # Mengecek ukuran dataframe
```

```python
proporsi = 0.7 # Variable pembagi
# df_1 bekerja dengan method sample yang mengambil baris secara acak sebesar proporsi yang ditentukan pada parameter frac.
# Lalu membentuk dataframe baru.
df_1 = df.sample(frac=proporsi) 
# df_2 bekerja dengan method drop yang menghapus baris sesuai jumlah tertentu.
df_2 = df.drop(df_1.index) 
'''
Opsi lain untuk membentuk dataframe serupa dengan df_2
df_3 = df.sample(frac=1-proporsi) 
'''

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_2.shape}')
#print(f'df_3 shape: {df_4.shape}')
```

# Pandas 10
## Mengganti nama (label) kolom pada Data Frame berdasarkan pola

```python
df2 = pd.read_csv('./DataSet/titanicfull.csv') # Membuat sebuah data frame dengan referensi dari CSV file.
# Mengganti nama kolom dari dataframe yang ingin dibuat.
df2.columns = ['Pclass', 'Survival status', 'full Name', 'Sex ', ' Age', 'Sib SP', 'Parch', 'Ticket', 'Fare','Cabin', 'Embarked']
df2_backup = df2.copy(deep=True)
# Menampilkan 5 baris pertama.
df2.head()
```

```python
"""
Mengganti nama kolom menggunakan method replace dan lower.
- Method lower berfungsi untuk mengubah setiap nama kolom menjadi lowercase (huruf kecil)
- Method replace berfungsi untuk mengubah suatu karakter menjadi karakter lainnya sesuai yang dicantumkan pada parameter.
"""
df2.columns = df2.columns.str.replace(' ', '_').str.lower()
df2.head()

# df2.columns = df2.columns.str.capitalize()
# Mengganti nama kolom menjadi kapital.
```

```python
df2 = df2_backup.copy(deep =True)
df2.columns = df2.columns.str.lower().str.strip().str.replace(' ','_')
# String strip akan membuang kelebihan spasi.
df2.head()
```

# Pandas 11
## Melakukan seleksi kolom dan baris pada data frame menggunakan loc

```python
n_rows = 10
n_cols = 5
cols = tuple("ABCDE")

df3 = pd.DataFrame(np.random.randint(1, 20, size=(n_rows,n_cols)), columns=cols)
df3
```

```python
'''
Melakukan seleksi kolom dan baris menggunakan loc.
parameter pertama dari loc digunakan untuk menentukan baris dataframe yang ingin ditampilkan.
parameter kedua digunakan untuk menentukan kolom dataframe yang ingin ditampilkan.
'''
df3.loc[[0,3,4], ['B','E']]
```

```python
# Melakukan seleksi baris dengan kondisi tertentu.
df3.loc[df3['B']>10, ['B','D','E']]
```

```python
# Slicing datframe dengan loc
# Berbeda dengan slicing pada list, jika menggunakan loc start index dan end index keduanya bersifat inklusif
df3.loc[0:4, 'B':'D']
```
